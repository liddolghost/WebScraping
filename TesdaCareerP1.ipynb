```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
edge_driver_path = 'C:\webdrivers\msedgedriver.exe'
service = webdriver.edge.service.Service(edge_driver_path)
driver = webdriver.Edge(service=service)

driver.get('https://tesda.gov.ph/Careers')
driver.maximize_window()
```

```python
#position
career_position = driver.find_elements(By.XPATH, '//tbody/tr/td[1]')

#item number
career_item_number = driver.find_elements(By.XPATH, '//tbody/tr/td[2]')

#office
career_office = driver.find_elements(By.XPATH, '//tbody/tr/td[3]')

#publication date
career_publication_date = driver.find_elements(By.XPATH, '//tbody/tr/td[4]')
```

```python
career_results = []  # This creates an empty list named 'career_results'
```

```python
for i in range(len(career_position)):
    career_dictionary={
        'Position':career_position[i].text,
        'Item Number':career_item_number[i].text,
        'Office':career_office[i].text,
        'Publication Date':career_publication_date[i].text}
    
    career_results.append(career_dictionary)
```

```python
dataframe_career = pd.DataFrame(career_results)
```

```python
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
edge_driver_path = 'C:\webdrivers\msedgedriver.exe'
service = webdriver.edge.service.Service(edge_driver_path)
driver = webdriver.Edge(service=service)

driver.get('https://doh.gov.ph/doh-budget')
driver.maximize_window() 
```

```python
#year
budget_year = driver.find_elements(By.XPATH, '//tbody/tr/td[1]')

#ps
budget_ps = driver.find_elements(By.XPATH, '//tbody/tr/td[2]')

#mooe
budget_mooe = driver.find_elements(By.XPATH, '//tbody/tr/td[3]')

#co
budget_co = driver.find_elements(By.XPATH, '//tbody/tr/td[4]')

#total
budget_total = driver.find_elements(By.XPATH, '//tbody/tr/td[5]')
```

```python
budget_results = []
```

```python
for i in range(len(budget_year)):
    budget_dictionary={
        'Year':budget_year[i].text,
        'PS':budget_ps[i].text,
        'MOOE':budget_mooe[i].text,
        'CO':budget_co[i].text,
        'Total':budget_total[i].text}
    
    budget_results.append(budget_dictionary)
```

```python
dataframe_budget = pd.DataFrame(budget_results)
```
